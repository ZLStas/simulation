```kotlin
import org.zhuravel.DelayType
import org.zhuravel.Simulation

data class SimulationParameters(
    val delayType: DelayType,
    val from: Int,
    val until: Int,
    val networkDegradation: Int,
    val ticks: Int,
    val ab: Pair<Int, Int>,
    val Tmax: Int,
    val Dme: Float,
    val Bc: Int
) {
    fun printRanges() {
        val Tme_min = Tmax * Dme + ab.first
        val Tme_max = Tmax + ab.second

        val Tle_min = Tmax * (1 - Dme) + ab.first
        val Tle_max = Tmax + Bc + ab.second

        println("Tme = ${Tme_min}...${Tme_max}")
        println("Tle = ${Tle_min}...${Tle_max}")
    }
}

val params = SimulationParameters(
    delayType = DelayType.LOG_NORMAL,
    from = 700,
    until = 3475,
    networkDegradation = 7000,
    ticks = 1000,
    ab = 1 to 10,
    Tmax = 6950,
    Dme = 0.5.toFloat(),
    Bc = 3000
)

fun runSimulationWithBrokenLinks(simParams: SimulationParameters, numberOfBrokenLinks: Int) =
    Simulation.runWithBrokenLeaderToNodeLinks(
        simParams.delayType,
        simParams.from,
        simParams.until,
        simParams.networkDegradation,
        simParams.ticks,
        simParams.ab,
        simParams.Tmax,
        simParams.Dme,
        simParams.Bc,
        numberOfBrokenLinks
    )

params.printRanges()

val avarageReconfigurationWithOneBrokenLink = runSimulationWithBrokenLinks(params, 1)
val avarageReconfigurationWithTwoBrokenLink = runSimulationWithBrokenLinks(params, 2)
val avarageReconfigurationWithTreeBrokenLink = runSimulationWithBrokenLinks(params, 3)

val avarageReconfigurationWithForeBrokenLink = runSimulationWithBrokenLinks(params, 4)

val avarageReconfigurationWithFiveBrokenLink = runSimulationWithBrokenLinks(params, 5)
//
//val avarageReconfigurationWithSixBrokenLink = runSimulationWithBrokenLinks(params, 6)
//val avarageReconfigurationWithSevenBrokenLink = runSimulationWithBrokenLinks(params, 7)

%useLatestDescriptors
%use kandy

plot {
    line {
        x(avarageReconfigurationWithOneBrokenLink.keys)
        y(avarageReconfigurationWithOneBrokenLink.values)
        color = Color.GREEN
    }
    line {
        x(avarageReconfigurationWithTwoBrokenLink.keys)
        y(avarageReconfigurationWithTwoBrokenLink.values)
        color = Color.LIGHT_GREEN
    }
    line {
        x(avarageReconfigurationWithTreeBrokenLink.keys)
        y(avarageReconfigurationWithTreeBrokenLink.values)
        color = Color.YELLOW
    }
    line {
        x(avarageReconfigurationWithForeBrokenLink.keys)
        y(avarageReconfigurationWithForeBrokenLink.values)
        color = Color.PURPLE
    }
    line {
        x(avarageReconfigurationWithFiveBrokenLink.keys)
        y(avarageReconfigurationWithFiveBrokenLink.values)
        color = Color.BLUE
    }
//    line {
//        x(avarageReconfigurationWithSixBrokenLink.keys)
//        y(avarageReconfigurationWithSixBrokenLink.values)
//        color = Color.PEACH
//    }
//    line {
//        x(avarageReconfigurationWithSevenBrokenLink.keys)
//        y(avarageReconfigurationWithSevenBrokenLink.values)
//        color = Color.ORANGE
//    }
}
```

```kotlin
import kotlin.random.Random
import kotlin.math.sqrt

fun simulateParameters(
    numNodes: Int = 5,
    maxTheta: Int = 100,
    iterations: Int = 1000,
    badNodes: List<Int> = listOf(1, 3) //    
) {
    val leaderCounts = IntArray(numNodes) { 0 }
    var optimalLeaderCounts = 0
    var badLeaderCounts = 0
    var reconfigCounts = 0

    repeat(iterations) {
        //     
        val delays = Array(numNodes) { IntArray(numNodes) { Random.nextInt(50, 300) } }
        for (i in 0 until numNodes) delays[i][i] = Int.MAX_VALUE //    '

        //  '     
        for (badNode in badNodes) {
            for (j in 0 until numNodes) {
                delays[badNode][j] += 200
                delays[j][badNode] += 200
            }
        }

        //    (    )
        val optimalNode = delays.indices.minByOrNull { delays[it].sum() } ?: 0

        //  -   
        val timeouts = IntArray(numNodes) {
            val sortedDelays = delays[it].sorted()
            val tDlcc = sortedDelays[1] //   
            val randomComponent = Random.nextInt(0, maxTheta)
            tDlcc + randomComponent
        }

        //   (   -)
        val leader = timeouts.indices.minByOrNull { timeouts[it] } ?: 0
        leaderCounts[leader]++

        //     ?
        if (leader == optimalNode) {
            optimalLeaderCounts++
        }

        //       ?
        if (badNodes.contains(leader)) {
            badLeaderCounts++
        }

        //   ?
        if (timeouts[leader] > maxTheta * 1.5) { //   
            reconfigCounts++
        }
    }

    // 1. CV -  
    val meanLeader = leaderCounts.average()
    val stdLeader = sqrt(leaderCounts.map { (it - meanLeader).toDouble().pow(2) }.sum() / leaderCounts.size)
    val cv = stdLeader / meanLeader

    // 2. P_bad_leader -    
    val pBadLeader = badLeaderCounts.toDouble() / iterations

    // 5. f_leader -      
    val fLeader = leaderCounts.map { it.toDouble() / iterations }

    // 7. f_reconfig -  
    val fReconfig = reconfigCounts.toDouble() / iterations

    // 8. P_optimal -    
    val pOptimal = optimalLeaderCounts.toDouble() / iterations

    //  
    println("1. CV ( ): %.2f".format(cv))
    println("2. P_bad_leader (   ): %.2f".format(pBadLeader))
    println("5.       (f_leader):")
    fLeader.forEachIndexed { node, freq ->
        val status = if (badNodes.contains(node)) " ( )" else ""
        println("    ${node + 1}: %.2f%%$status".format(freq * 100))
    }
    println("7. f_reconfig ( ): %.2f%%".format(fReconfig * 100))
    println("8. P_optimal (   ): %.2f%%".format(pOptimal * 100))
}

fun main() {
    //  
    val numNodes = 5
    val maxTheta = 100
    val iterations = 1000
    val badNodes = listOf(1, 3) //    

    //  
    simulateParameters(numNodes, maxTheta, iterations, badNodes)
}
```
